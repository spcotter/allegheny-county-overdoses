```python
import urllib


base_url = "https://data.wprdc.org/dataset/7fb0505e-8e2c-4825-b22c-4fbee8fc8010/resource/"
urllib.urlretrieve ( base_url + "2d963e35-4f69-495e-985e-55acd72c87ca/download/crimelabaccidentaldrugdeathsextract2017.csv", "crimelabaccidentaldrugdeathsextract2017.csv")
urllib.urlretrieve ( base_url + "386ca0ed-717d-46d3-8ed2-da37553ec6d1/download/crimelabaccidentaldrugdeathsextract2016.csv", "crimelabaccidentaldrugdeathsextract2016.csv")
urllib.urlretrieve ( base_url + "502136a8-649b-44c5-9b08-ec2733201e5f/download/crimelabaccidentaldrugdeathsextract2015.csv", "crimelabaccidentaldrugdeathsextract2015.csv")
urllib.urlretrieve ( base_url + "7146df41-9531-4327-961a-362ad6f3d2c3/download/crimelabaccidentaldrugdeathsextract2014.csv", "crimelabaccidentaldrugdeathsextract2014.csv")
urllib.urlretrieve ( base_url + "ee648fab-3048-48b6-9757-f8dfbae378c8/download/crimelabaccidentaldrugdeathsextract2013.csv", "crimelabaccidentaldrugdeathsextract2013.csv")
```

```python
import pandas

frames = []
frames.append( pandas.read_csv("crimelabaccidentaldrugdeathsextract2017.csv") )
frames.append( pandas.read_csv("crimelabaccidentaldrugdeathsextract2016.csv") )
frames.append( pandas.read_csv("crimelabaccidentaldrugdeathsextract2015.csv") )
frames.append( pandas.read_csv("crimelabaccidentaldrugdeathsextract2014.csv") )
frames.append( pandas.read_csv("crimelabaccidentaldrugdeathsextract2013.csv") )

data = pandas.concat(frames)

data
```

```python

data['Death Date Time'] = pandas.to_datetime(data['Death Date'] + ' ' + data['Death Time'])
```

```python
data['Case Year'].value_counts()
```

```python
all_drugs = data['Combined OD1']  \
    .append(data['Combined OD2']) \
    .append(data['Combined OD3']) \
    .append(data['Combined OD4']) \
    .append(data['Combined OD5']) \
    .append(data['Combined OD6']) \
    .append(data['Combined OD7'])
all_drugs.value_counts() / data['Combined OD1'].size
```

```python
data['heroin'] = (data['Combined OD1'] == 'Heroin') | \
    (data['Combined OD2'] == 'Heroin') | \
    (data['Combined OD3'] == 'Heroin') | \
    (data['Combined OD4'] == 'Heroin') | \
    (data['Combined OD5'] == 'Heroin') | \
    (data['Combined OD6'] == 'Heroin') | \
    (data['Combined OD7'] == 'Heroin')

data['heroin'].value_counts()
```

```python
data['fentanyl'] = (data['Combined OD1'] == 'Fentanyl') | \
    (data['Combined OD2'] == 'Fentanyl') | \
    (data['Combined OD3'] == 'Fentanyl') | \
    (data['Combined OD4'] == 'Fentanyl') | \
    (data['Combined OD5'] == 'Fentanyl') | \
    (data['Combined OD6'] == 'Fentanyl') | \
    (data['Combined OD7'] == 'Fentanyl')

data['fentanyl'].value_counts()
```

```python
pandas.crosstab(data['heroin'], data['fentanyl'])
```

```python
data['heroin_or_fentanyl'] = data['heroin'] | data['fentanyl']
data['heroin_or_fentanyl'].value_counts()
```

```python
incident_zip_x_od = pandas.crosstab(data['Incident Zip'], data['heroin_or_fentanyl'], margins=True)
incident_zip_x_od.sort_values(True, ascending=False)
```

```python
decedent_zip_x_od = pandas.crosstab(data['Decedent Zip'], data['heroin_or_fentanyl'], margins=True)
decedent_zip_x_od.sort_values(True, ascending=False)
```
